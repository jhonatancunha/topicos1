```python
import numpy as np
import pandas as pd
import base64
import codecs
```

```python
#pegar mensagem, mensagem completa, nome do repositorio onde o commit foi feito
raw = pd.read_csv("data2.csv")
data = raw.drop(["Unnamed: 8", "Unnamed: 9"], axis=1)
data
```

```python
def decode_msg(message):
    try:
        return base64.b64decode(message[2:-1:]).decode('utf-8')
    except:
        return message[2:-1:]


data['message'] = data['message'].apply(lambda row: decode_msg(row))
data
```

```python
def get_extension(row):
  return row.split(".")[-1]

data['extension'] = data['filename'].apply(lambda row: get_extension(row))
data
```

```python
def teste(row):
  a = row[2::].split()
  result = []

  for i in a:
    result.append(i[2:-1:])

  return " ".join(result)

data["repo"] = data["repo"].apply(lambda row: teste(row))
```

```python
author = data['author'].drop_duplicates().to_list()
author
```

```python
contributions_from_one_author = data.loc[data['author'] == author[50]]
contributions_from_one_author.sort_values(by="authored_at", ascending=True)
```
