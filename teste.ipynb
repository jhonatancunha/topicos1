```python
import numpy as np
import pandas as pd
import base64
import codecs
import matplotlib.pyplot as plt
from tqdm import tqdm
from matplotlib import colors
import matplotlib.patches as mpatches



%matplotlib inline
```

```python
data = pd.read_csv("data2.csv")
data
```

```python
def decode_msg(message):
    try:
        return base64.b64decode(message[2:-1:]).decode('utf-8')
    except:
        return message[2:-1:]


data['message'] = data['message'].apply(lambda row: decode_msg(row))
data
```

```python
def get_extension(row):
  return row.split(".")[-1]

data['extension'] = data['filename'].apply(lambda row: get_extension(row))
data
```

```python
def teste(row):
  a = row[2::].split()
  result = []

  for i in a:
    result.append(i[2:-1:])

  return " ".join(result)

data["repo"] = data["repo"].apply(lambda row: teste(row))
data
```

```python
language_targets = ['java', 'js', 'ts', 'py', 'c', 'go', 'php', 'cpp', 'cs', 'r', 'rb', 'swift', 'rs', 'rlib']
language_colors = {'java': 1, 'js': 2, 'ts': 3, 'py': 4, 'c': 5, 'go': 6, 'php': 7, 'cpp': 8, 'cs': 9, 'r': 10, 'rb': 11, 'swift': 12, 'rs': 13, 'rlib': 14}
```

```python
authors = data['author'].drop_duplicates()
authors
```

### Obtendo autores que e a quantidade de contribuÃ§Ãµes que fizeram

```python
df_author_filtered = pd.DataFrame()

for name in authors:
  contributions = data.loc[data['author'] == name]
  shas = contributions['sha'].drop_duplicates()
  qtd_commits = len(shas)

  df_author_filtered = df_author_filtered.append(pd.DataFrame({"name": [name], "qtd_commits": [qtd_commits]}))

display(df_author_filtered)
```

### Filtrando autores com mais de 8 contribuiÃ§Ãµes e menos de 16753


```python
#pegar autores entre com mais de 8 commits
author_more_8_commits = df_author_filtered[df_author_filtered['qtd_commits'] >= 8]
author_more_8_commits = author_more_8_commits[author_more_8_commits['qtd_commits'] < 16753]
author_more_8_commits = author_more_8_commits.sort_values(by="qtd_commits", ascending=False)


display(author_more_8_commits)
name_authors_more_8_commits = author_more_8_commits['name'].to_list()
qtd_name_authors_more_8_commits = len(name_authors_more_8_commits)
```

```python
df_authors = pd.DataFrame()


for name in tqdm(name_authors_more_8_commits):
    contributions_from_one_author = data.loc[data['author'] == name]
    contributions_from_one_author = contributions_from_one_author[contributions_from_one_author["extension"].isin(language_targets)]
    shas = contributions_from_one_author['sha'].drop_duplicates().to_list()


    for idx, sha in enumerate(shas):
        contributions_from_one_sha = contributions_from_one_author.loc[contributions_from_one_author['sha'] == sha]
        language = contributions_from_one_sha["extension"].value_counts().index.tolist()[0]

        aux = pd.DataFrame({"language": [language], "author": [name], "commit": idx})        
        
        df_authors = df_authors.append(aux)
    

df_authors

df_authors.to_csv("./contributions.csv")
```

### Obtendo maior numero de contribuiÃ§Ãµes

```python
maximo = 0

for name in tqdm(name_authors_more_8_commits):
  qtd_contrib = len(df_authors.loc[df_authors['author'] == name])
  
  maximo = qtd_contrib if maximo < qtd_contrib else maximo 


print(maximo)
df_authors
```

```python

data = np.zeros(shape=(qtd_name_authors_more_8_commits, maximo))

for idx, name in enumerate(name_authors_more_8_commits):
    author_contrib = df_authors.loc[df_authors['author'] == name]
    for _, contrib in author_contrib.iterrows():
        commit = contrib['commit']
        language = contrib['language']
        data[idx][commit] = language_colors[language]


display(data)
```

```python
list_of_colors = ['white', 'red', 'blue', 'green', 'orange', 'yellow', 'pink', 'cyan', 'purple', 'black', 'brown', 'thistle', 'slategrey', 'lime', 'olive']
cmap = colors.ListedColormap(list_of_colors)

for lang, key in language_colors.items():
    print("%s - %s" % (lang, list_of_colors[key]))

fig = plt.figure(figsize=(100, 50))
plt.imshow(data, cmap=cmap)

# draw gridlines
graph_line = plt.plot(np.array([0, maximo]), np.array([0, qtd_name_authors_more_8_commits]))

legend_list = []
valid_list_of_colors = list_of_colors[1::]

for idx, color in enumerate(valid_list_of_colors):
    legend_list.append(mpatches.Patch(color=color, label=language_targets[idx]))

    
line = graph_line.pop(0)
line.remove()
    
plt.legend(handles=legend_list, fontsize=30)

plt.savefig('graph.png')
plt.show()

##trocar cor do python
```
