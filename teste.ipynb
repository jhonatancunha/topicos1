```python
import numpy as np
import pandas as pd
import base64
import codecs
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
#pegar mensagem, mensagem completa, nome do repositorio onde o commit foi feito
data = pd.read_csv("data2.csv")
data
```

```python
def decode_msg(message):
    try:
        return base64.b64decode(message[2:-1:]).decode('utf-8')
    except:
        return message[2:-1:]


data['message'] = data['message'].apply(lambda row: decode_msg(row))
data
```

```python
def get_extension(row):
  return row.split(".")[-1]

data['extension'] = data['filename'].apply(lambda row: get_extension(row))
data
```

```python
def teste(row):
  a = row[2::].split()
  result = []

  for i in a:
    result.append(i[2:-1:])

  return " ".join(result)

data["repo"] = data["repo"].apply(lambda row: teste(row))
```

```python
data
```

```python
authors = data['author'].drop_duplicates().to_list()
print(len(authors))
```

```python
# date_with_time = data.drop_duplicates(subset=["authored_at"], keep='first')['authored_at'].drop_duplicates().to_list()
# years = list(set([i.split(" ")[0] for i in date_with_time]))

count = [i for i in range(0, len(authors))]
```

```python
df_authors = pd.DataFrame()

for name in tqdm(authors):
    contributions_from_one_author = data.loc[data['author'] == name].drop_duplicates(subset=["sha"], keep='first')
    sorted_contributions_from_one_author = contributions_from_one_author.sort_values(by="authored_at", ascending=True)
    amount_of_commits = sorted_contributions_from_one_author.shape[0]
        
    df_authors = df_authors.append(pd.DataFrame([[name, amount_of_commits]], columns=['author','qtd_commits']),ignore_index = True)
```

```python
df_authors.sort_values(by="qtd_commits", ascending=False)
```

```python
#pegar autores entre 35 e 117 commits
author_35_117_commits = df_authors[df_authors['qtd_commits'] <= 16753]
author_35_117_commits = author_35_117_commits[author_35_117_commits['qtd_commits'] >= 8]
author_35_117_commits.describe()
```
