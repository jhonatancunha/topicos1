```python
not_dummy_projects = pd.read_csv("not_dummy_repo_500_att.csv", header=None)[0].values.tolist()
amount_of_not_dummy_projects = len(not_dummy_projects)
amount_of_not_dummy_projects
```

```python
data = pd.read_csv("formatted_data.csv")
data
```

```python
data_matrix = np.zeros((amount_of_not_dummy_projects, amount_of_not_dummy_projects))
matrix_of_projects = pd.DataFrame(data=data_matrix, index=not_dummy_projects, columns=not_dummy_projects)
matrix_of_projects
```

```python
not_dummy_projects

for ndp in tqdm(not_dummy_projects):
    authors_from_not_dummy_projects =  data.loc[data["repo"].str.contains(ndp, regex=False)]['author'].drop_duplicates()
    
    for author in authors_from_not_dummy_projects:
        # obtendo todos os commits dos autores do projeto nao dummy 
        contribs_from_this_author = data.loc[data['author'] == author]
        
        # obtendo todos os commits que nÃ£o sao do projeto 'ndp'
        contribs_from_this_author = contribs_from_this_author.loc[contribs_from_this_author['repo'].str.contains(ndp, regex=False) == False]

        repos_and_qtd_commits = contribs_from_this_author.groupby(['repo'])['repo'].count()
        
        
        for repo in repos_and_qtd_commits.index:
            repo_to_search = repo.split()[0].split("_")
            
            if len(repo_to_search) == 1:
                repo_to_search = repo_to_search[0]
            else:
                repo_to_search = repo_to_search[1]
            
            if repo_to_search in not_dummy_projects:
                # print(ndp, repo_to_search)
                matrix_of_projects[ndp][repo_to_search] += 1
                
        # break
            
            
        # obtendo repositorios que estÃ£o na lista dos que possuem mais de 500 commits
        # contribs_from_this_author = contribs_from_this_author.loc[contribs_from_this_author['repo'].str.contains('|'.join(not_dummy_projects))]
        
        # display('|'.join(not_dummy_projects))
        # display(contribs_from_this_author)
        
        # contando quantos commits existem de cada repo
        # qtd_commits_by_repo = contribs_from_this_author.groupby(['repo'])['repo'].count()
        
        # break
        # for repo in qtd_commits_by_repo.index:
            # qtd_commits = qtd_commits_by_repo[repo]
            # matrix_of_projects[ndp][repo] = qtd_commits
```
